ques - 6 Explain any three AI-powered ETL (Extract, Transform, Load) tools and how they are used in analytics.

- AI-powered ETL tools such as Talend, Informatica, and Azure Data Factory automate data preparation, improve data quality, and reduce processing time, making analytics more efficient, accurate, and scalable.

- One widely used AI-powered ETL tool is Talend Data Fabric. Talend uses machine learning to automatically detect data quality issues such as missing values, duplicates, and inconsistent formats. It also suggests data transformations and cleansing rules based on patterns it identifies in the data.

Where to use - In analytics, Talend helps ensure that clean and reliable data is loaded into data warehouses, which improves the accuracy of dashboards, reports, and predictive models.

- Another important AI-powered ETL tool is Informatica Intelligent Data Management Cloud (IDMC). This tool uses an AI engine called CLAIRE to automate data mapping, schema matching, and anomaly detection. Informatica reduces manual effort by learning from previous transformations and recommending optimal ETL workflows.

Where to use - In analytics, it helps organizations integrate data from multiple sources efficiently, enabling faster insights and better decision-making.

- A third AI-powered ETL tool is Microsoft Azure Data Factory. Azure Data Factory incorporates AI capabilities for automated data profiling, anomaly detection, and integration with Azure Machine Learning services. It allows analysts to build intelligent data pipelines that can adapt to changes in data structure.

Where to use - In analytics, Azure Data Factory is commonly used to move large volumes of data into cloud data warehouses where it can be analyzed using tools like Power BI and Azure Synapse Analytics.

Ques - 7 Give three applications of Data Analytics across different industries and explain how it creates value.
- Data Analytics is in the retail and e-commerce industry. Retail companies analyze customer purchase history, browsing behavior, and feedback data to understand customer preferences. This helps in demand forecasting, personalized recommendations, and inventory management. By using data analytics, retailers reduce stock shortages and overstocking, improve customer satisfaction, and increase sales through targeted marketing and personalized offers.

- Another important application of Data Analytics is in the healthcare industry. Hospitals and healthcare providers analyze patient records, medical histories, and treatment outcomes to improve diagnosis and patient care. Analytics is used to predict disease outbreaks, reduce hospital readmission rates, and optimize resource allocation such as staff and equipment.

- A third application of Data Analytics is in the finance and banking industry. Banks analyze transaction data, customer behavior, and credit histories to detect fraud, assess credit risk, and personalize financial products. Real-time analytics helps identify suspicious activities and prevent financial losses.


Ques - 8 Explain the evolution of analytics, highlighting the different stages from Descriptive to Generative AI.

- Analytics has evolved from Descriptive analytics, which explains what happened, to Diagnostic analytics that identifies why it happened. Predictive analytics forecasts future outcomes using historical data, while Prescriptive analytics recommends optimal actions. Cognitive analytics enables systems to understand and reason like humans using AI. The latest stage, Generative AI, creates new content, insights, and solutions autonomously.
